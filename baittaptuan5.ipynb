```python

from math import isqrt
def CheckPrime(n):
    for i in range(2,isqrt(n)+1):
        if n % i == 0:
            return False
    return True

def main():
    n = int(input())
    print(CheckPrime(n))
if __name__ == "__main__":
    main()

```

```python
#W5A4 VIết hàm kiểm tra một số có phải số hoàn hảo hay không, xử lý trên dữ liệu kiểu số nguyên.
def check_perfect_number(num):
    sum_of_divisors = 0
    for i in range(1,num):
        if num % i == 0:
            sum_of_divisors += i
    return sum_of_divisors == num

```

```python
#W5A5 Viết hàm đưa vào một list số nguyên và một số nguyên dương k. Hãy tìm và trả về vị trí của
#phần tử đầu tiên có giá trị k trong list số nguyên, nếu không có thì trả về -1

def find_first_occurrence(int_list, k):
    for index, value in enumerate(int_list):
        if value == k:
            return index
    return -1
```

```python
#W5A6 
def factorial(n):
    for i in range(1,n):
        n = n*i
    return n 

factorial(5)
```

```python
#W5A7

def calculator(num1, num2, operate):
    if operate == '-':
        return num1 - num2
    if operate == '+':
        return num1 + num2
    if operate == '/':
        if num2 == 0:
            return False
        return num1 / num2
    if operate == '*':
        return num1*num2
    
calculator(4,5,'-')
    
    
```

```python
#W5A8

def hamming_distance(x:int,y:int) -> int:
    xor_result = x^y
    distance = xor_result.bit_count()
    return distance
x,y = map(int,input().split())
print(hamming_distance(x,y))
```

```python
def dang_cau(a,b):
    mp1 ={}
    mp2 = {}
    for i in range(len(a)):
        if a[i] in mp1 and mp1[a[i]] != b[i]:
            return False
        if b[i] in mp2 and mp2[b[i]] != a[i]:
            return False
        mp1[a[i]] = b[i]
        mp2[b[i]] = a[i]
    return True
a,b = input().split()
print(dang_cau(a,b))
```

```python
"""Viết hàm kiểm tra một số có phải số Amstrong hay không? Hàm trả về kiểu bool, có tham số là một số nguyên x"""

def check_amstrong(n):
    n_str = str(n)
    count = len(n_str)
    total = 0 
    for i in n_str:
        total += int(i)**(count)

    return total == n


check_amstrong(372)
```

```python
"""
Viết hàm kiêểm tra một số có phải số chính phương hay không? Hàm trả về kiểu bool, có tham số là một số nguyên x
"""
from math import sqrt
def check_square(n):
    i = int(sqrt(n))
    return i*i == n
check_square(26)
```

```python
"""
VIết hàm kiểm tra một số có phải số đối xứng không, xử lý trên dữ liệu kiểu số nguyên.
"""
def check_palindrome(n):
    res = list(str(n))
    return res == res[::-1]
    
check_palindrome(123)
```
