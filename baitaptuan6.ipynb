```python
"""
Nhập vào một dòng gồm các số nguyên. Hãy loại bỏ phần tử trùng nhưng
giữ nguyên thứ tự xuất hiện đầu tiên, rồi in ra list kết quả.
Yêu cầu in ra đúng dạng list Python, ví dụ [1, 2, 3]
"""

nums = list(map(int, input().split()))
res = []
for i in nums:
    if i not in res:
        res.append(i)
print(res)
```

```python
"""
Nhập một dòng số nguyên. In ra list tổng tích lũy (phần tử i là tổng từ đầu
tới vị trí i), giữ nguyên thứ tự
"""

nums = list(map(int, input().split()))
res = []
total = 0
for i in nums:
    total += i
res.append(total)
print(res)
```

```python
"""
Nhập một tuple số nguyên (một dòng) và số nguyên k (dòng kế). In ra tuple
sau khi xoay trái k vị trí. (Giữ nguyên thứ tự tương đối sau xoay; không sắp
xếp.)
"""
nums = tuple(map(int,input().split()))
k = int(input())
if len(nums) > 0:
    k = k%len(nums)
    rotate = nums[k:] + nums[:k]
else:
    rotate = nums
print(rotate)

```

```python
# Nhập một dòng gồm các cặp key:value cách nhau bởi dấu cách. Tạo
# dictionary ánh xạ mỗi key -> list các value theo thứ tự xuất hiện. Nếu key
# lặp lại, thêm vào list của key đó

pairs = input().split()
dict_res = {}
for pair in pairs:
    key, value = pair.split(':')
    if key not in dict_res:
        dict_res[key] = []
    dict_res[key].append(value)
print(dict_res)
```

```python
"""
Nhập một list số nguyên, In ra dictionary {'positives': ..., 'negatives': ...,
'zeros': ...} (đếm số >0, <0, =0)
"""
n = list(map(int, input().split()))
dictionary = {
    'positives': sum(1 for i in n if i > 0),
    'negatives': sum(1 for i in n if i < 0),
    'zeros': sum(1 for i in n if i == 0)
}
print(dictionary)
```

```python
"""
Viết chương trình nhập vào một tuple, in ra phần tử đầu, phần tử cuối, và
tuple đảo ngược
"""
nums = tuple(map(int,input().split()))
dau = nums[0]
cuoi = nums[-1]
dao_nguoc = nums [::-1]

print(dau, cuoi, dao_nguoc)
```

```python
"""Viết chương trình nhập vào một list các chuỗi, sau đó đếm số lần xuất hiện
của từng chuỗi và lưu vào dictionary"""
words = input().split()
freg = {}
for i in words:
    if i in freg:
        freg[i] += 1
    else:
        freg[i] = 1
print(freg)
```

```python
"""Viết chương trình nhập vào hai dictionary. Sau đó gộp lại thành một
dictionary mới trong đó nếu một key xuất hiện trong cả hai dict, giá trị sẽ
được cộng lại, nếu chỉ có trong một dict thì giữ nguyên. In ra dictionary mới
theo thứ tự từ điển tăng dần của key."""
#W6A9
d1 = eval(input())
d2 = eval(input())
d = {}
for key, value in d1.items():
    d[key] = value
for key,value in d2.items():
    if key in d:
        d[key] += value
    else:
        d[key] = value
result = sorted(d)
print(d)
    

```

```python
"""
Viết chương trình nhập vào một danh sách các số nguyên, và số nguyên
dương k. In ra list gồm các cặp số khác nhau có tổng bằng k. Mỗi cặp là
một tuple, hai số trong tuple có thể giống nhau, thứ tự xuất hiện trong danh
sách của số thứ nhất nhỏ hơn thứ tự của số thứ hai. Các tuple trong list được 
sắp xếp theo thứ tự tăng dần của phần tử đầu tiên trong tuple.
"""

nums = list(map(int,input().split()))
k = int(input())
pairs = []
nums.sort()
for i in range(len(nums)):
    for j in range(i+1,len(nums)):
        if nums[i] + nums[j] == k:
            pairs.append((nums[i],nums[j]))
print(pairs)
```

```python
"""
[Tách số chẵn và lẻ từ tuple]
Nhập vào một tuple chứa các số nguyên. Hãy tạo ra hai tuple mới: một chứa
tất cả các số chẵn và một chứa tất cả các số lẻ từ tuple ban đầu, giữ nguyên
thứ tự tương đối. In ra hai tuple này trên hai dòng riêng biệt.
"""

nums = tuple(map(int,input().split()))
chan = tuple(n for n in nums if n % 2 == 0)
le = tuple(n for n in nums if n % 2 == 1)
print(chan)
print(le)


```

```python
"""[Tìm phần tử xuất hiện nhiều nhất]
Nhập vào một list các số nguyên. Hãy tìm số xuất hiện nhiều lần nhất trong
list. Nếu có nhiều số cùng có số lần xuất hiện nhiều nhất, hãy in ra số nhỏ
nhất trong số đó."""

nums = list(map(int,input().split()))
freg = {}
for i in nums:
    if i in freg:
        freg[i] += 1
    else:
        freg[i] = 1
max_frequency = 0 
result_number = float('inf')

for key,value in freg.items():
    if value > max_frequency:
        max_frequency = value
        result_number = key
    elif value == max_frequency:
        if key < result_number:
            result_number = key
print(result_number)

    
        
```

```python
"""[Đảo ngược key và value của dictionary]
Nhập vào một dictionary có value là duy nhất (không trùng nhau). Hãy tạo
một dictionary mới bằng cách đảo ngược key và value của dictionary ban
đầu. In ra dictionary mới."""

dict1 = eval(input())
reverse_dict = {}
for key,value in dict1.items():
    reverse_dict[value] = key
print(reverse_dict)
    
```

```python
"""[Giao của hai danh sách]
Nhập vào hai dòng, mỗi dòng là một danh sách các số nguyên. Hãy tìm các
phần tử chung (giao) của hai danh sách này và in ra một list mới chứa các
phần tử chung đó, không trùng lặp."""

list1 = list(map(int,input().split()))
list2 = list(map(int,input().split()))
list_chung = []
for i in list1:
    if i in list2:
        list_chung.append(i)
print(list_chung)

```

```python
"""[Lọc dictionary theo giá trị]
Nhập vào một dictionary có key là chuỗi và value là số nguyên, và một số
nguyên k ở dòng tiếp theo. Hãy tạo một dictionary mới chỉ chứa các cặp
key-value từ dictionary ban đầu mà có value lớn hơn k. In ra dictionary
mới."""

d = eval(input())
k = int(input())

filtered = {}
for key,value in d.items():
    if value > k:
        filtered[key] = value
print(filtered)
```
